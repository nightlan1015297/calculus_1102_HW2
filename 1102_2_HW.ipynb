```python
import matplotlib.pyplot as plt
import math
sun_mass = 1.989e30
G = 6.67e-11

def acc(r_square):
    return G*sun_mass/r_square
```

```python
initial_x = 1.47e11
initial_y = 0
initial_vx = 0
initial_vy = 3.03e4
x = [initial_x]
y = [initial_y]
vx = [initial_vx]
vy = [initial_vy]
```

```python
orbit_period = 60*60*24*365
dt = 60
step = int(orbit_period/dt)

for i in range(step):
    r_square = x[i]**2+y[i]**2
    a = acc(r_square)
    r = math.sqrt(r_square)
    ax =- a * x[i]/r
    ay =- a * y[i]/r
    vx.append(vx[i] + ax*dt)
    vy.append(vy[i] + ay*dt)
    x.append(x[i] + vx[i]*dt)
    y.append(y[i] + vy[i]*dt)

plt.plot(x,y)
plt.show()
```
